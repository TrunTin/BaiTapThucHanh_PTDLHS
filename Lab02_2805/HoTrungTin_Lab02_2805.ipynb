```python
print("=====================================================================")
print("PHẦN 1: THỐNG KÊ DỮ LIỆU")
print("=====================================================================")

import pandas as pd
import numpy as np

df = pd.read_csv('processed_dulieuxettuyendaihoc.csv')

# Yêu cầu 1: Sắp xếp dữ liệu điểm DH1 theo thứ tự tăng dần
print("\n" + "="*70)
print("YÊU CẦU 1: SẮP XẾP DỮ LIỆU ĐIỂM DH1 THEO THỨ TỰ TĂNG DẦN")
print("="*70)

df_dh1_sorted = df[['STT', 'DH1', 'GT', 'KT', 'KV']].sort_values('DH1', ascending=True).reset_index(drop=True)
print("\nDữ liệu sắp xếp DH1 tăng dần:")
print(df_dh1_sorted)
```

```python
# Yêu cầu 2: Sắp xếp dữ liệu điểm DH2 tăng dần theo nhóm giới tính
print("\n" + "="*70)
print("YÊU CẦU 2: SẮP XẾP ĐIỂM DH2 TĂNG DẦN THEO NHÓM GIỚI TÍNH")
print("="*70)

df_dh2_sorted_by_gt = df[['STT', 'DH2', 'GT', 'KT', 'KV']].sort_values(['GT', 'DH2'], ascending=[True, True]).reset_index(drop=True)
print("\nDữ liệu sắp xếp DH2 tăng dần theo giới tính (F trước, M sau):")
print(df_dh2_sorted_by_gt)

print("\n" + "-"*70)
print("CHI TIẾT THỐNG KÊ THEO GIỚI TÍNH:")
print("-"*70)

for gt in ['F', 'M']:
    df_gt = df_dh2_sorted_by_gt[df_dh2_sorted_by_gt['GT'] == gt]
    print(f"\nGiới tính {gt} (Tổng: {len(df_gt)} học sinh):")
    print(df_gt.to_string(index=False))
```

```python
import pandas as pd
import numpy as np

print("\n" + "="*70)
print("YÊU CẦU 3: PIVOT TABLE THỐNG KÊ DH1 THEO KHỐI THI (KT)")
print("="*70)

# Tạo pivot table bằng danh sách hàm chuẩn
pivot_dh1_kt = pd.pivot_table(
    df,
    values='DH1',
    index='KT',
    aggfunc=[
        'count', 
        'sum', 
        'mean', 
        'median', 
        'min', 
        'max', 
        'std', 
        lambda x: np.percentile(x, 25),
        lambda x: np.percentile(x, 50),
        lambda x: np.percentile(x, 75)
    ]
)

pivot_dh1_kt.columns = ['Count', 'Sum', 'Mean', 'Median', 'Min', 'Max', 'Std', 'Q1', 'Q2', 'Q3']

print("\nBảng Pivot Table DH1 theo KT:")
print(pivot_dh1_kt)

print("\n" + "-"*70)
print("BẢNG CHI TIẾT THỐNG KÊ DH1 THEO KHỐI THI:")
print("-"*70)

for kt in df['KT'].unique():
    df_kt = df[df['KT'] == kt]['DH1']
    print(f"\nKhối {kt}:")
    print(f"   Count (Số lượng):      {len(df_kt)}")
    print(f"   Sum (Tổng):            {df_kt.sum():.4f}")
    print(f"   Mean (Trung bình):     {df_kt.mean():.4f}")
    print(f"   Median (Trung vị):     {df_kt.median():.4f}")
    print(f"   Min (Giá trị nhỏ nhất): {df_kt.min():.4f}")
    print(f"   Max (Giá trị lớn nhất): {df_kt.max():.4f}")
    print(f"   Std (Độ lệch chuẩn):   {df_kt.std():.4f}")
    print(f"   Q1 (Phân vị 25%):      {np.percentile(df_kt, 25):.4f}")
    print(f"   Q2 (Phân vị 50%):      {np.percentile(df_kt, 50):.4f}")
    print(f"   Q3 (Phân vị 75%):      {np.percentile(df_kt, 75):.4f}")
```

```python
import pandas as pd
import numpy as np

print("\n" + "="*70)
print("YÊU CẦU 4: PIVOT TABLE THỐNG KÊ DH1 THEO KHỐI THI (KT) VÀ KHU VỰC (KV)")
print("="*70)

# Tạo pivot table với danh sách hàm chuẩn (không dùng Tuple tự đặt tên)
pivot_dh1_kt_kv = pd.pivot_table(
    df,
    values='DH1',
    index=['KT', 'KV'],
    aggfunc=[
        'count', 
        'sum', 
        'mean', 
        'median', 
        'min', 
        'max', 
        'std', 
        lambda x: np.percentile(x, 25),
        lambda x: np.percentile(x, 50),
        lambda x: np.percentile(x, 75)
    ]
)

pivot_dh1_kt_kv.columns = ['Count', 'Sum', 'Mean', 'Median', 'Min', 'Max', 'Std', 'Q1', 'Q2', 'Q3']

print("\nBảng Pivot Table DH1 theo KT và KV:")
print(pivot_dh1_kt_kv)

print("\n" + "-"*70)
print("BẢNG CHI TIẾT THỐNG KÊ DH1 THEO KHỐI THI VÀ KHU VỰC:")
print("-"*70)

for kt in sorted(df['KT'].unique()):
    print(f"\n{'='*70}")
    print(f"KHỐI {kt}:")
    print(f"{'='*70}")
    
    for kv in sorted(df[df['KT'] == kt]['KV'].unique()):
        df_subset = df[(df['KT'] == kt) & (df['KV'] == kv)]['DH1']
        
        print(f"\n   Khu vực {kv}:")
        print(f"    Count (Số lượng):      {len(df_subset)}")
        print(f"    Sum (Tổng):            {df_subset.sum():.4f}")
        print(f"    Mean (Trung bình):     {df_subset.mean():.4f}")
        print(f"    Median (Trung vị):     {df_subset.median():.4f}")
        print(f"    Min (Giá trị nhỏ nhất): {df_subset.min():.4f}")
        print(f"    Max (Giá trị lớn nhất): {df_subset.max():.4f}")
        print(f"    Std (Độ lệch chuẩn):   {df_subset.std():.4f}")
        
        if len(df_subset) > 0:
            print(f"    Q1 (Phân vị 25%):      {np.percentile(df_subset, 25):.4f}")
            print(f"    Q2 (Phân vị 50%):      {np.percentile(df_subset, 50):.4f}")
            print(f"    Q3 (Phân vị 75%):      {np.percentile(df_subset, 75):.4f}")
        else:
            print("    Q1 (Phân vị 25%):      NaN")
            print("    Q2 (Phân vị 50%):      NaN")
            print("    Q3 (Phân vị 75%):      NaN")

print("\n" + "="*70)
print("Hoàn thành phân tích dữ liệu - PHẦN 1!")
print("="*70)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Đọc dữ liệu
df = pd.read_csv('processed_dulieuxettuyendaihoc.csv')

print("\n" + "="*70)
print("PHẦN 3: TRỰC QUAN HÓA DỮ LIỆU THEO NHÓM PHÂN LOẠI")
print("="*70)

# Định nghĩa bảng màu nhất quán cho kết quả Đậu/Rớt (KQXT) để biểu đồ trực quan hơn
# 0: Rớt (Đỏ nhạt), 1: Đậu (Xanh lá)
colors_kqxt = ['#E64B35', '#4DBBD5'] 

# -------------------------------------------------------------------------
# 1. Trực quan dữ liệu học sinh nữ trên các nhóm XL1, XL2, XL3 dạng unstacked
# -------------------------------------------------------------------------
print("\nĐang xử lý Yêu cầu 1...")
female_df = df[df['GT'] == 'F']

# Gom dữ liệu đếm tần số xếp loại cho từng năm (Lớp 10, 11, 12)
xl1_counts = female_df['XL1'].value_counts()
xl2_counts = female_df['XL2'].value_counts()
xl3_counts = female_df['XL3'].value_counts()

# Tạo DataFrame gộp với các hàng là nhóm [Y, TB, K, G, XS]
xl_categories = ['Y', 'TB', 'K', 'G', 'XS']
xl_df = pd.DataFrame(index=xl_categories)
xl_df['XL1'] = xl_df.index.map(xl1_counts).fillna(0)
xl_df['XL2'] = xl_df.index.map(xl2_counts).fillna(0)
xl_df['XL3'] = xl_df.index.map(xl3_counts).fillna(0)

# Xoay bảng để Ox thể hiện XL1, XL2, XL3 và màu sắc thể hiện nhóm xếp loại (Unstacked)
xl_df_transposed = xl_df.T

# Vẽ biểu đồ
ax1 = xl_df_transposed.plot(kind='bar', figsize=(9, 5), color=['#E64B35', '#4DBBD5', '#00A087', '#3C5488', '#F39C12'])
ax1.set_title('Xếp loại học lực của học sinh Nữ qua các năm (XL1, XL2, XL3)')
ax1.set_xlabel('Các năm học (Lớp 10, Lớp 11, Lớp 12)')
ax1.set_ylabel('Số lượng học sinh')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.legend(title='Xếp loại')
plt.show()


# -------------------------------------------------------------------------
# 2. Trực quan dữ liệu KQXT trên nhóm học sinh có khối thi A, A1, B thuộc khu vực 1, 2
# -------------------------------------------------------------------------
print("Đang xử lý Yêu cầu 2...")
# Lọc theo khối thi và khu vực (Lưu ý: KV trong dữ liệu có thể là số hoặc chuỗi, ta dùng ép kiểu hoặc so sánh chuỗi)
df['KV_str'] = df['KV'].astype(str).str.replace('.0', '', regex=False)
subset_yc2 = df[df['KT'].isin(['A', 'A1', 'B']) & df['KV_str'].isin(['1', '2'])]

# Lập bảng chéo giữa Khối thi + Khu vực và Kết quả xét tuyển
yc2_crosstab = pd.crosstab(index=[subset_yc2['KT'], subset_yc2['KV_str']], columns=subset_yc2['KQXT'])

ax2 = yc2_crosstab.plot(kind='bar', figsize=(10, 5), color=colors_kqxt)
ax2.set_title('Kết quả xét tuyển (KQXT) theo Khối thi và Khu vực (1, 2)')
ax2.set_xlabel('(Khối thi, Khu vực)')
ax2.set_ylabel('Số lượng thí sinh')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.legend(title='KQXT', labels=['Rớt (0)', 'Đậu (1)'])
plt.tight_layout()
plt.show()


# -------------------------------------------------------------------------
# 3. Trực quan dữ liệu số lượng thí sinh từng khu vực dựa trên từng nhóm khối thi
# -------------------------------------------------------------------------
print("Đang xử lý Yêu cầu 3...")
kt_kv_crosstab = pd.crosstab(index=df['KT'], columns=df['KV_str'])

ax3 = kt_kv_crosstab.plot(kind='bar', figsize=(9, 5))
ax3.set_title('Số lượng thí sinh từng Khu vực dựa trên từng nhóm Khối thi')
ax3.set_xlabel('Khối thi')
ax3.set_ylabel('Số lượng thí sinh')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.legend(title='Khu vực')
plt.show()


# -------------------------------------------------------------------------
# 4. Trực quan dữ liệu số lượng thí sinh đậu, rớt trên từng nhóm khối thi
# -------------------------------------------------------------------------
print("Đang xử lý Yêu cầu 4...")
kt_kqxt_crosstab = pd.crosstab(index=df['KT'], columns=df['KQXT'])

ax4 = kt_kqxt_crosstab.plot(kind='bar', figsize=(9, 5), color=colors_kqxt)
ax4.set_title('Số lượng thí sinh Đậu/Rớt trên từng nhóm Khối thi')
ax4.set_xlabel('Khối thi')
ax4.set_ylabel('Số lượng thí sinh')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.legend(title='KQXT', labels=['Rớt (0)', 'Đậu (1)'])
plt.show()


# -------------------------------------------------------------------------
# 5. Trực quan dữ liệu số lượng thí sinh đậu rớt trên từng nhóm khu vực
# -------------------------------------------------------------------------
print("Đang xử lý Yêu cầu 5...")
kv_kqxt_crosstab = pd.crosstab(index=df['KV_str'], columns=df['KQXT'])

ax5 = kv_kqxt_crosstab.plot(kind='bar', figsize=(9, 5), color=colors_kqxt)
ax5.set_title('Số lượng thí sinh Đậu/Rớt trên từng nhóm Khu vực')
ax5.set_xlabel('Khu vực')
ax5.set_ylabel('Số lượng thí sinh')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.legend(title='KQXT', labels=['Rớt (0)', 'Đậu (1)'])
plt.show()


# -------------------------------------------------------------------------
# 6. Trực quan dữ liệu số lượng thí sinh đậu rớt dựa trên từng nhóm dân tộc
# -------------------------------------------------------------------------
print("Đang xử lý Yêu cầu 6...")
# Xử lý chuẩn hóa tên dân tộc (Kinh hoặc số mã hóa)
df['DT_str'] = df['DT'].fillna(0).astype(str).str.replace('.0', '', regex=False)
dt_kqxt_crosstab = pd.crosstab(index=df['DT_str'], columns=df['KQXT'])

ax6 = dt_kqxt_crosstab.plot(kind='bar', figsize=(9, 5), color=colors_kqxt)
ax6.set_title('Số lượng thí sinh Đậu/Rớt dựa trên từng nhóm Dân tộc')
ax6.set_xlabel('Dân tộc (0 là Kinh)')
ax6.set_ylabel('Số lượng thí sinh')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.legend(title='KQXT', labels=['Rớt (0)', 'Đậu (1)'])
plt.show()


# -------------------------------------------------------------------------
# 7. Trực quan dữ liệu số lượng thí sinh đậu rớt dựa trên từng nhóm giới tính
# -------------------------------------------------------------------------
print("Đang xử lý Yêu cầu 7...")
gt_kqxt_crosstab = pd.crosstab(index=df['GT'], columns=df['KQXT'])

ax7 = gt_kqxt_crosstab.plot(kind='bar', figsize=(7, 5), color=colors_kqxt)
ax7.set_title('Số lượng thí sinh Đậu/Rớt dựa trên từng nhóm Giới tính')
ax7.set_xlabel('Giới tính (F: Nữ, M: Nam)')
ax7.set_ylabel('Số lượng thí sinh')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.legend(title='KQXT', labels=['Rớt (0)', 'Đậu (1)'])
plt.show()

print("\n" + "="*70)
print("Hoàn thành trực quan hóa dữ liệu - PHẦN 3!")
print("="*70)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

print("\n" + "="*70)
print("PHẦN 4: PHÂN LOẠI VÀ TRỰC QUAN HÓA NÂNG CAO")
print("="*70)

# Đọc dữ liệu từ file CSV
df = pd.read_csv('processed_dulieuxettuyendaihoc.csv')
df['STT'] = df.index + 1

# -------------------------------------------------------------------------
# 2. Tạo biến phân loại phanlopt1 cho T1
# -------------------------------------------------------------------------
# Định nghĩa các khoảng điểm [0->5), [5->7), [7->8), [8->10]
bins = [0, 5, 7, 8, 10.1] 
# Sửa nhãn: Sử dụng 'k' (kém), 'tb' (trung bình), 'kh' (khá), 'g' (giỏi)
labels = ['k', 'tb', 'kh', 'g'] 

df['phanlopt1'] = pd.cut(df['T1'], bins=bins, labels=labels, right=False)

print("\nBảng phân loại phanlopt1 cho T1 (10 hàng đầu):")
print(df[['STT', 'T1', 'phanlopt1']].head(10))

# -------------------------------------------------------------------------
# 3. Lập bảng tần số cho biến phanlopt1
# -------------------------------------------------------------------------
phanlopt1_freq = df['phanlopt1'].value_counts().reindex(labels, fill_value=0)
print("\nBảng tần số biến phanlopt1:")
print(phanlopt1_freq)

# Bảng màu đại diện cho 4 nhóm phân loại
colors = {'k': '#E64B35', 'tb': '#4DBBD5', 'kh': '#00A087', 'g': '#F39C12'}
labels_vn = {'k': 'Kém (k)', 'tb': 'Trung bình (tb)', 'kh': 'Khá (kh)', 'g': 'Giỏi (g)'}

# -------------------------------------------------------------------------
# 4. Vẽ biểu đồ đường Multiple Line cho biến T1 được phân loại bởi biến phanlopt1
# -------------------------------------------------------------------------
print("\nVẽ biểu đồ Multiple Line cho T1 theo phân loại phanlopt1...")
fig, ax = plt.subplots(figsize=(14, 6))

# Để các đường không bị đứt gãy do chứa giá trị rỗng (NaN), 
# ta lọc bỏ các dòng NaN của từng nhóm trước khi vẽ nối điểm nối đường.
for cat in labels:
    subset = df[df['phanlopt1'] == cat].sort_values('STT')
    if not subset.empty:
        # Vẽ đường nối các học sinh cùng nhóm lại với nhau
        ax.plot(subset['STT'], subset['T1'], marker='o', linestyle='-', 
                color=colors[cat], label=labels_vn[cat], markersize=5, linewidth=1.5)

ax.set_title('Biểu đồ Multiple Line - Biến động điểm T1 theo từng nhóm Phân loại', fontsize=14, fontweight='bold', pad=15)
ax.set_xlabel('Số thứ tự học sinh (STT)', fontsize=12)
ax.set_ylabel('Điểm môn Toán (T1)', fontsize=12)
ax.set_xlim(1, len(df))
ax.set_ylim(0, 11)
ax.set_xticks(range(1, len(df) + 1, 5))
ax.legend(title='Phân loại học lực T1', loc='upper right')
ax.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

# -------------------------------------------------------------------------
# 5. Vẽ biểu đồ Drop-line cho biến T1 được phân loại bởi biến phanlopt1
# -------------------------------------------------------------------------
print("\nVẽ biểu đồ Drop-line cho T1 theo phân loại phanlopt1...")
fig, ax = plt.subplots(figsize=(14, 6))

# Vẽ các đường Drop-line hạ thẳng góc từ điểm số xuống trục hoành
for cat in labels:
    subset = df[df['phanlopt1'] == cat]
    # Vẽ các cột sọc vlines
    ax.vlines(subset['STT'], ymin=0, ymax=subset['T1'], colors=colors[cat], alpha=0.4, linewidth=1)
    # Vẽ các điểm chấm tròn ở đỉnh đầu để định vị điểm số
    ax.plot(subset['STT'], subset['T1'], marker='o', linestyle='None', 
            color=colors[cat], label=labels_vn[cat], markersize=5)

ax.set_title('Biểu đồ Drop-line (Stem Plot) của điểm T1 theo nhóm Phân loại', fontsize=14, fontweight='bold', pad=15)
ax.set_xlabel('Số thứ tự học sinh (STT)', fontsize=12)
ax.set_ylabel('Điểm môn Toán (T1)', fontsize=12)
ax.set_xlim(1, len(df))
ax.set_ylim(0, 11)
ax.set_xticks(range(1, len(df) + 1, 5))
ax.legend(title='Phân loại học lực T1')
ax.grid(True, linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()

print("\n" + "="*70)
print("Hoàn thành xử lý và trực quan nâng cao - PHẦN 4!")
print("="*70)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis, probplot

print("\n" + "="*70)
print("PHẦN 5: MÔ TẢ DỮ LIỆU VÀ KHẢO SÁT DẠNG PHÂN PHỐI")
print("="*70)

# Đọc dữ liệu gốc
df = pd.read_csv('processed_dulieuxettuyendaihoc.csv')

# Đảm bảo tạo biến phân lớp T1 đồng nhất bài trước
bins_t1 = [0, 5, 7, 8, 10.1]
labels_t1 = ['k', 'tb', 'kh', 'g']
df['phanlopT1'] = pd.cut(df['T1'], bins=bins_t1, labels=labels_t1, right=False)

# Chuẩn hóa cột khu vực thành chuỗi sạch để tránh lỗi TypeError khi sort
df['KV_clean'] = df['KV'].astype(str).str.replace('.0', '', regex=False)

# =========================================================================
# 1. Mô tả và khảo sát phân phối cho biến T1
# =========================================================================
print("\n--- 1. MÔ TẢ VÀ KHẢO SÁT PHÂN PHỐI CHO BIẾN T1 ---")

T1 = df['T1'].dropna()

# Tính toán các chỉ số thống kê mô tả hướng tâm và phân tán
T1_count = len(T1)
T1_mean = T1.mean()
T1_median = T1.median()
T1_std = T1.std()
T1_var = T1.var()
T1_min = T1.min()
T1_max = T1.max()
T1_q1 = np.percentile(T1, 25)
T1_q3 = np.percentile(T1, 75)
T1_iqr = T1_q3 - T1_q1

# Xác định chuẩn 10 đại lượng biểu diễn trong và xung quanh Box-Plot
lower_whisker = max(T1_min, T1_q1 - 1.5 * T1_iqr)
upper_whisker = min(T1_max, T1_q3 + 1.5 * T1_iqr)
lower_outliers = T1[T1 < lower_whisker].tolist()
upper_outliers = T1[T1 > upper_whisker].tolist()

T1_skew = skew(T1, bias=False)
T1_kurt = kurtosis(T1, fisher=True, bias=False)

print("\n[+] Bảng 10 đại lượng đặc trưng khảo sát dữ liệu của Box-Plot:")
print(f"1. Minimum (Giá trị nhỏ nhất)         : {T1_min:.2f}")
print(f"2. Lower Whisker (Biên dưới hạn định) : {lower_whisker:.2f}")
print(f"3. Q1 (Tứ phân vị thứ nhất - 25%)     : {T1_q1:.2f}")
print(f"4. Median (Trung vị - 50%)            : {T1_median:.2f}")
print(f"5. Q3 (Tứ phân vị thứ ba - 75%)       : {T1_q3:.2f}")
print(f"6. Upper Whisker (Biên trên hạn định) : {upper_whisker:.2f}")
print(f"7. Maximum (Giá trị lớn nhất)         : {T1_max:.2f}")
print(f"8. IQR (Khoảng tứ phân vị)            : {T1_iqr:.2f}")
print(f"9. Lower Outliers (Dị biệt dưới)      : {lower_outliers}")
print(f"10. Upper Outliers (Dị biệt trên)     : {upper_outliers}")

print("\n[+] Chỉ số hình dáng phân phối:")
print(f"- Hệ số lệch (Skewness) : {T1_skew:.3f}")
print(f"- Hệ số nhọn (Kurtosis) : {T1_kurt:.3f} (Excess Kurtosis)")

# Trực quan hóa Box-Plot
fig, ax = plt.subplots(figsize=(10, 3))
ax.boxplot(T1, vert=False, patch_artist=True,
           boxprops=dict(facecolor='#6FA8DC', color='#3C78D8'),
           medianprops=dict(color='#E97451', linewidth=2),
           whiskerprops=dict(color='#3C78D8', linewidth=1.5),
           capprops=dict(color='#3C78D8', linewidth=1.5),
           flierprops=dict(marker='o', markerfacecolor='#E64B35', markersize=6, alpha=0.8))
ax.set_title('Biểu đồ Box-Plot của biến T1', fontsize=12, fontweight='bold')
ax.set_xlabel('Điểm số T1')
ax.set_yticks([])
ax.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

# Trực quan hóa Histogram
fig, ax = plt.subplots(figsize=(10, 5))
ax.hist(T1, bins=10, color='#4DBBD5', edgecolor='black', alpha=0.8, density=False)
ax.set_title('Biểu đồ Histogram thể hiện tần suất phân phối T1', fontsize=12, fontweight='bold')
ax.set_xlabel('Điểm số T1')
ax.set_ylabel('Số lượng học sinh (Tần số)')
ax.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

# Vẽ biểu đồ kiểm chứng phân phối chuẩn QQ-Plot
fig, ax = plt.subplots(figsize=(7, 5))
probplot(T1, dist='norm', plot=ax)
ax.set_title('Biểu đồ QQ-Plot kiểm chứng phân phối chuẩn của T1', fontsize=12, fontweight='bold')
plt.tight_layout()
plt.show()

# Nhận xét phân phối T1
skew_desc = "lệch trái (đuôi kéo dài về phía điểm thấp)" if T1_skew < -0.1 else ("lệch phải" if T1_skew > 0.1 else "gần như đối xứng")
kurt_desc = "dẹt hơn phân phối chuẩn (platykurtic)" if T1_kurt < 0 else "nhọn hơn phân phối chuẩn (leptokurtic)"
print(f"\n[Nhận xét]: Biến T1 có giá trị trung bình ({T1_mean:.2f}) gần bằng trung vị ({T1_median:.2f}). "
      f"Hệ số Skewness = {T1_skew:.3f} cho thấy {skew_desc}. "
      f"Hệ số Kurtosis = {T1_kurt:.3f} chứng minh phân phối {kurt_desc}. "
      f"Dựa trên đường thẳng QQ-Plot, các điểm dữ liệu phân bố sát đường tuyến tính tiêu chuẩn, chứng tỏ T1 hội tụ xấp xỉ phân phối chuẩn tốt.")


# =========================================================================
# 2. Mô tả và khảo sát phân phối cho biến T1 trên từng nhóm phân lớp
# =========================================================================
print("\n--- 2. MÔ TẢ PHÂN PHỐI T1 TRÊN TỪNG NHÓM PHANLOPT1 ---")

# Vẽ Box-plot theo phân nhóm
fig, ax = plt.subplots(figsize=(10, 5))
df.boxplot(column='T1', by='phanlopT1', ax=ax, grid=False,
           patch_artist=True,
           boxprops=dict(facecolor='#AED6F1', color='#1F618D'),
           medianprops=dict(color='#641E16', linewidth=2))
ax.set_title('Biểu đồ Box-Plot của T1 phân rã theo nhóm phân lớp')
ax.set_xlabel('Nhóm học lực Toán (phanlopT1)')
ax.set_ylabel('Điểm môn Toán T1')
plt.suptitle('')
plt.tight_layout()
plt.show()

# Vẽ Histogram theo phân nhóm
fig, ax = plt.subplots(figsize=(10, 5))
colors_sub = {'k': '#E64B35', 'tb': '#4DBBD5', 'kh': '#00A087', 'g': '#F39C12'}
for cat in labels_t1:
    subset = df[df['phanlopT1'] == cat]['T1'].dropna()
    if len(subset) > 0:
        ax.hist(subset, bins=5, alpha=0.5, label=f'Nhóm {cat}', color=colors_sub[cat], edgecolor='black')
ax.set_title('Biểu đồ Histogram chồng xếp lớp của T1 theo nhóm phanlopT1')
ax.set_xlabel('Điểm T1')
ax.set_ylabel('Tần số số lượng')
ax.legend()
plt.tight_layout()
plt.show()

# Vẽ QQ-Plot đa khung hình cho từng nhóm học lực
fig, axs = plt.subplots(2, 2, figsize=(11, 9))
axs = axs.flatten()
for idx, cat in enumerate(labels_t1):
    subset = df[df['phanlopT1'] == cat]['T1'].dropna()
    if len(subset) > 1:
        probplot(subset, dist='norm', plot=axs[idx])
        axs[idx].set_title(f'QQ-Plot - Nhóm {cat}')
    else:
        axs[idx].text(0.5, 0.5, 'Dữ liệu quá ít', ha='center', va='center')
plt.tight_layout()
plt.show()


# =========================================================================
# 3. Khảo sát tương quan giữa biến DH1 theo biến T1
# =========================================================================
print("\n--- 3. KHẢO SÁT TƯƠNG QUAN GIỮA BIẾN DH1 THEO BIẾN T1 ---")

covariance_t1_dh1 = df['T1'].cov(df['DH1'])
correlation_t1_dh1 = df['T1'].corr(df['DH1'])

print(f"[+] Hiệp phương sai Covariance(T1, DH1) : {covariance_t1_dh1:.3f}")
print(f"[+] Hệ số tương quan Correlation(T1, DH1): {correlation_t1_dh1:.3f}")

fig, ax = plt.subplots(figsize=(8, 5))
ax.scatter(df['T1'], df['DH1'], alpha=0.75, color='#4DBBD5', edgecolor='black', s=50)
ax.set_title('Biểu đồ phân tán Scatter Plot giữa điểm thi DH1 theo điểm học bạ T1', fontsize=11, fontweight='bold')
ax.set_xlabel('Điểm học bạ môn Toán (T1)')
ax.set_ylabel('Điểm thi Đại học môn thứ 1 (DH1)')
ax.grid(True, linestyle='--', alpha=0.5)
# Chèn nhãn hiển thị tương quan
ax.text(0.05, 0.92, f'r = {correlation_t1_dh1:.3f}\nCov = {covariance_t1_dh1:.3f}', 
        transform=ax.transAxes, fontsize=11, bbox=dict(facecolor='white', alpha=0.8, edgecolor='gray'))
plt.tight_layout()
plt.show()


# =========================================================================
# 4. Hãy khảo sát tương quan giữa biến DH1 theo biến T1 trên từng nhóm khu vực
# =========================================================================
print("\n--- 4. KHẢO SÁT TƯƠNG QUAN DH1 THEO T1 TRÊN TỪNG NHÓM KHU VỰC ---")

# Lấy danh sách khu vực đã làm sạch và sắp xếp đúng bảng mã ký tự chuỗi
list_kv = sorted(df['KV_clean'].dropna().unique())

fig, axs = plt.subplots(2, 2, figsize=(12, 10))
axs = axs.flatten()

for idx, kv in enumerate(list_kv):
    sub_kv = df[df['KV_clean'] == kv]
    
    # Tính tương quan cục bộ trong từng khu vực
    cov_kv = sub_kv['T1'].cov(sub_kv['DH1'])
    corr_kv = sub_kv['T1'].corr(sub_kv['DH1'])
    
    axs[idx].scatter(sub_kv['T1'], sub_kv['DH1'], alpha=0.8, color='#00A087', edgecolor='black', s=45)
    axs[idx].set_title(f'Khu vực: {kv} (r = {corr_kv:.3f})', fontweight='bold')
    axs[idx].set_xlabel('Điểm T1')
    axs[idx].set_ylabel('Điểm DH1')
    axs[idx].grid(True, linestyle='--', alpha=0.4)
    axs[idx].text(0.05, 0.88, f'Cov = {cov_kv:.3f}', transform=axs[idx].transAxes, 
                  fontsize=9, bbox=dict(facecolor='white', alpha=0.7))

# Ẩn ô biểu đồ thừa thứ 4 nếu danh sách khu vực chỉ có 3 nhóm (1, 2, 2NT)
if len(list_kv) < 4:
    axs[3].axis('off')

plt.tight_layout()
plt.show()


# =========================================================================
# 5. Hãy khảo sát tương quan giữa các biến DH1, DH2, DH3
# =========================================================================
print("\n--- 5. KHẢO SÁT TƯƠNG QUAN GIỮA CÁC BIẾN DH1, DH2, DH3 ---")

dh_cols = ['DH1', 'DH2', 'DH3']
cov_matrix_dh = df[dh_cols].cov()
corr_matrix_dh = df[dh_cols].corr()

print("\n[+] Ma trận hiệp phương sai (Covariance Matrix):")
print(cov_matrix_dh.round(4))
print("\n[+] Ma trận hệ số tương quan (Correlation Matrix):")
print(corr_matrix_dh.round(4))

# Vẽ Scatter Matrix ma trận phân tán đồng bộ bằng matplotlib subplot phối hợp công thức kết đôi
fig, axs = plt.subplots(3, 3, figsize=(12, 12))
for i, row_var in enumerate(dh_cols):
    for j, col_var in enumerate(dh_cols):
        ax = axs[i, j]
        if i == j:
            # Đường chéo chính hiển thị biểu đồ phân phối Histogram đại lượng
            ax.hist(df[row_var].dropna(), bins=8, color='#4DBBD5', edgecolor='black', alpha=0.75)
        else:
            # Các ô ngoài đường chéo chính hiển thị tương quan điểm cặp
            ax.scatter(df[col_var], df[row_var], alpha=0.7, color='#E64B35', edgecolor='black', s=30)
            corr_val = df[col_var].corr(df[row_var])
            ax.text(0.1, 0.85, f'r = {corr_val:.2f}', transform=ax.transAxes, 
                    fontsize=10, fontweight='bold', bbox=dict(facecolor='white', alpha=0.7))
        
        # Thiết lập tên nhãn biên rìa ngoài
        if i == 2: ax.set_xlabel(col_var, fontweight='bold')
        if j == 0: ax.set_ylabel(row_var, fontweight='bold')
        ax.grid(True, linestyle='--', alpha=0.3)

plt.suptitle('Ma trận tương quan phân tán Scatter Matrix giữa DH1, DH2 và DH3', fontsize=14, fontweight='bold', y=0.98)
plt.tight_layout()
plt.show()

print("\n" + "="*70)
print("HOÀN THÀNH TOÀN BỘ KHẢO SÁT PHÂN PHỐI VÀ TƯƠNG QUAN - PHẦN 5!")
print("="*70)
```
