```python
# 2. Xem tên các tập tin có trong gói 'gutenberg'
gb = nltk.corpus.gutenberg
print("Gutenberg files : ", gb.fileids())
print("-" * 50)

# Truy cập tập tin Macbeth
macbeth = nltk.corpus.gutenberg.words('shakespeare-macbeth.txt')

# Hiển thị chiều dài văn bản (Đã sửa: Bọc print để không bị ẩn kết quả)
print("Chiều dài của văn bản Macbeth:", len(macbeth))

# Hiển thị 10 từ đầu tiên (Đã sửa: Bọc print để không bị ẩn kết quả)
print("10 từ đầu tiên:", macbeth[:10])
print("-" * 50)

# Trích 5 câu đầu tiên của tập tin
macbeth_sents = nltk.corpus.gutenberg.sents('shakespeare-macbeth.txt')
print("5 câu đầu tiên của tập tin:")
for i, sent in enumerate(macbeth_sents[:5], 1):
    print(f"Câu {i}: {' '.join(sent)}")
```

```python
# 3. Tìm từ với NLTK
text = nltk.Text(macbeth)

print("--- Các ngữ cảnh xuất hiện của từ 'Stage' ---")
text.concordance('Stage')

print("\n--- Các từ xuất hiện trước và sau từ 'Stage' ---")
text.common_contexts(['Stage'])

print("\n--- Các từ có cách sử dụng tương tự từ 'Stage' ---")
text.similar('Stage')
```

```python
# 4. Phân tích tần số của các từ
import string

fd = nltk.FreqDist(macbeth)
print("10 từ thông dụng nhất ban đầu:", fd.most_common(10))

# Xem các stopword trong tiếng Anh (Đã loại bỏ dòng download thừa)
sw = set(nltk.corpus.stopwords.words('english'))
print(f"\nSố lượng stopwords tiếng Anh: {len(sw)}")
print("10 stopwords ví dụ:", list(sw)[:10])

# Loại bỏ các từ stopword trong biến macbeth
macbeth_filtered = [w for w in macbeth if w.lower() not in sw]
print(f"\nChiều dài sau khi lọc stopword: {len(macbeth_filtered)}")

fd = nltk.FreqDist(macbeth_filtered)
print("10 từ thông dụng sau khi lọc stopword:", fd.most_common(10))

# Loại bỏ thêm các dấu câu
punctuation = set(string.punctuation)
macbeth_filtered2 = [w.lower() for w in macbeth if w.lower() not in sw and w.lower() not in punctuation]

fd = nltk.FreqDist(macbeth_filtered2)
print("\n10 từ thông dụng sau khi lọc stopword và dấu câu:", fd.most_common(10))
```

```python
# 5. Lựa chọn các từ đặc biệt trong văn bản
long_words = [w for w in macbeth if len(w) > 12]
print("--- Các từ có độ dài lớn hơn 12 (đã sắp xếp) ---")
print(sorted(long_words))

# Rút trích các từ có chứa chuỗi ‘ious’
ious_words = [w for w in macbeth if 'ious' in w]
ious_words = set(ious_words)
print("\n--- Các từ có chứa chuỗi 'ious' (đã loại trùng) ---")
print(sorted(ious_words))
```

```python
# 6. Bigrams và collocations
bgrms = nltk.FreqDist(nltk.bigrams(macbeth_filtered2))
print("--- 15 Bigrams phổ biến nhất ---")
print(bgrms.most_common(15))
print("-" * 50)

# Ngoài bigram ra, còn có trigram, sự kết hợp của 3 từ
tgrms = nltk.FreqDist(nltk.trigrams(macbeth_filtered2))
print("--- 10 Trigrams phổ biến nhất ---")
print(tgrms.most_common(10))
```

```python
# 7. Sử dụng văn bản trên mạng
import nltk
from urllib import request

# Sử dụng https để tránh lỗi tự động chuyển hướng dữ liệu
url = "https://www.gutenberg.org/files/2554/2554-0.txt"

try:
    # CÁCH 1: Giả lập trình duyệt Chrome để không bị chặn + Cài đặt thời gian chờ timeout = 30 giây
    req = request.Request(url, headers={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64)'})
    response = request.urlopen(req, timeout=30)
    raw = response.read().decode('utf8')

    print("--- 75 ký tự đầu tiên của văn bản thô ---")
    print(raw[:75])
    print("\n" + "-" * 50 + "\n")

    # Tách từ (Tokenize) và chuyển đổi sang đối tượng Text
    tokens = nltk.word_tokenize(raw)
    webtext = nltk.Text(tokens)

    print("--- 12 tokens đầu tiên sau khi xử lý ---")
    print(webtext[:12])

except Exception as e:
    print(f"[CẢNH BÁO] Không thể kết nối đến URL Project Gutenberg do nghẽn mạng: {e}")
    print("Vui lòng kiểm tra lại đường truyền mạng hoặc liên hệ để nhận giải pháp dự phòng!")
```

```python
# 8. Rút trích văn bản từ trang html
import nltk 
from urllib import request
from bs4 import BeautifulSoup

url = "http://news.bbc.co.uk/2/hi/health/2284783.stm"

# Giả lập User-Agent trình duyệt để tránh lỗi hệ thống chặn (HTTP 403)
req = request.Request(url, headers={'User-Agent': 'Mozilla/5.0'})
html = request.urlopen(req).read().decode('utf8')

print("--- 120 ký tự HTML đầu tiên ---")
print(html[:120])
print("\n" + "-" * 50 + "\n")

# Dùng BeautifulSoup bóc tách text thuần
raw = BeautifulSoup(html, "lxml").get_text()

# Tách từ và cấu trúc hóa đối tượng Text
tokens = nltk.word_tokenize(raw)
text = nltk.Text(tokens)

print(f"Số lượng tokens trích xuất được: {len(text)}")
print("10 tokens đầu tiên:", text[:10])
```

```python
# 8. Phân tích cảm xúc người dùng
nltk.download('movie_reviews')
import random

reviews = nltk.corpus.movie_reviews
documents = [
    (list(reviews.words(fileid)), category)
    for category in reviews.categories()
    for fileid in reviews.fileids(category)
]

# Trộn dữ liệu ngẫu nhiên
random.seed(42)  # Thêm seed để đảm bảo kết quả giống nhau giữa các lần chạy
random.shuffle(documents)

# Xem nội dung review đầu tiên (dòng 0, cột 0)
first_review = ' '.join(documents[0][0])
print("--- Nội dung review đầu tiên ---")
print(first_review[:500] + "...")  # In 500 ký tự đầu cho đỡ dài màn hình
print("\n--- Nhãn cảm xúc đầu tiên ---")
print(documents[0][1])

# Tạo bảng phân phối tần số các từ trong corpus
all_words = nltk.FreqDist(w.lower() for w in reviews.words())

# TỐI ƯU: Chỉ lấy 2000 từ phổ biến nhất làm tính năng (giúp chạy cực nhanh)
word_features = list(all_words)[:2000]

def document_features(document, word_features):
    document_words = set(document)
    features = {}
    for word in word_features:
        features['{}'.format(word)] = (word in document_words)
    return features

# Định nghĩa hàm document_features(), tạo 1 tập các documents
print("\nĐang trích xuất đặc trưng (quá trình này mất vài giây)...")
featuresets = [(document_features(d, word_features), c) for (d, c) in documents]
print("Số lượng mẫu đặc trưng:", len(featuresets))

# Tạo tập train và tập test chuẩn (1500 dòng đầu cho train, 500 dòng còn lại cho test)
train_set, test_set = featuresets[:1500], featuresets[1500:]

# Huấn luyện mô hình bằng thuật toán Naïve Bayes
print("Đang huấn luyện mô hình Naive Bayes...")
classifier = nltk.NaiveBayesClassifier.train(train_set)

# Tính toán và hiển thị độ chính xác
accuracy = nltk.classify.accuracy(classifier, test_set)
print(f"\nĐộ chính xác của mô hình: {accuracy * 100:.2f}%")

# Hiển thị 10 đặc trưng mang tính quyết định nhất
print("\n--- 10 đặc trưng quan trọng nhất ---")
classifier.show_most_informative_features(10)
```
